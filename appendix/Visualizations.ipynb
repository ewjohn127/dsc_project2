# Data Visualizations for Presentation

```python
# imports
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# reading in the data to a dataframe
df = pd.read_csv('data/cleaned_data.csv')
df.drop(['id','date','Unnamed: 0'],axis=1,inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
# scatterplot
# sqft_living vs price
plt.figure(figsize=(9,9))
sns.set_style('dark')
sns.scatterplot(x=df['sqft_living'],y=df['price'], alpha=0.5)
plt.title('Sqft Living and Price', fontsize=18)
plt.xlabel('Sqft. Living', fontsize=15)
plt.ylabel('Price', fontsize=15)
```

## Barchart for Type of Communities

```python
# creating a dataframe for only Seattle locations
seattle = df[(df['suburb'] == 0) & (df['rural'] == 0)]
seattle
```

```python
# creating a dataframe for only suburban locations
suburb = df[df['suburb'] == 1]
suburb
```

```python
rural = df[df['rural'] == 1]
rural
```

```python
# tick labels
x_tick_labels = ['Seattle','Suburban','Rural']

# average prices
avg_prices = [seattle['price'].mean(), suburb['price'].mean(), rural['price'].mean()]
avg_prices
```

```python
# creating the barplot
plt.figure(figsize=(7,7))
sns.barplot(x=x_tick_labels, y=avg_prices, edgecolor='black', palette='light:#5A9')
plt.title('Average Home Prices by Community Type', fontsize=20)
plt.xlabel('Community Type', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.ylabel('Price ($)', fontsize=16)
plt.savefig('avgprice_bar')
```
